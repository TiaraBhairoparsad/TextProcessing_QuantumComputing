# What is text processing?
>- About text processing: 
- According to Rouse, M. 2017., text processing refers to the techniques and methodologies used to analyse, transform, and manipulate textual data, which is often unstructured and requires special handling to extract meaningful information, derive insights, and perform various analyses.
- Utilizing natural language processing (NLP) and machine learning, subfields of artificial intelligence, text processing tools can automatically comprehend human language and derive valuable insights from textual data. 
- Text processing involves the automated handling of electronic text, encompassing tasks such as creating, modifying, searching, replacing, and formatting content, as well as generating refined reports.
- It utilizes computer commands to systematically manipulate textual data and focuses on handling text at a high level within computing, dealing with the automated transmission of information. 
-  Unlike algorithms, text processing can be seen as a series of sequentially executed macros that are simpler in nature, incorporating filtering techniques and pattern-action expressions to manage text. It employs text processing utilities rather than text editing tools, operating in a sequential manner rather than providing random access to text. 
- Working directly at the presentation layer and indirectly impacting the application layer, text processing deals with raw data and remains independent of proprietary formats and techniques.
- Text processing involves not only breaking down textual data into tokens but also understanding the context in which these tokens appear. This contextual understanding is crucial for tasks such as sentiment analysis, where the sentiment of a sentence can change based on the surrounding words and phrases. Techniques like named entity recognition (NER) and part-of-speech tagging further enhance context-aware processing. 
- Beyond stemming and lemmatisation, text processing also includes normalization tasks like handling uppercase and lowercase letters consistently. This ensures that the same word or phrase is treated uniformly throughout the text, which is essential for accurate analysis and modeling.
- In addition to removing stopwords, text processing often includes techniques to handle noisy data such as typos, abbreviations, slang, and non-standard language forms. Techniques like spell-checking, entity disambiguation, and normalization of acronyms and abbreviations can improve data quality.
- Text processing techniques may vary significantly depending on the language of the text and the specific domain or industry context. For instance, tokenisation and analysis of medical texts may require specialized dictionaries and terminology, while legal texts may demand specific handling of jargon and legal terms.
- Text processing is closely integrated with machine learning techniques in NLP. Preprocessed textual data serves as input to machine learning algorithms for tasks such as classification, clustering, sequence prediction (e.g., in language modeling), and text generation. Techniques such as word embeddings (e.g., Word2Vec, GloVe) enable machines to understand semantic relationships between words.
- Processing text data also involves ethical considerations such as privacy, bias, and fairness, especially when dealing with sensitive information or making decisions based on textual analysis. Ensuring fairness in automated decision-making processes and protecting user privacy are critical aspects of responsible text processing.
- With the advent of big data, text processing techniques have evolved to handle large volumes of text efficiently. Distributed processing frameworks (e.g., Apache Hadoop, Apache Spark) and algorithms (e.g., MapReduce) are employed to scale text processing tasks across distributed computing environments, enabling real-time analysis of massive textual datasets.

>- According to Anon. 2020., the key components of text processing include:
- Tokenisation: The process of breaking down text into individual units called tokens, which can be words, phrases, or symbols. It facilitates further processing like frequency analysis, syntactic analysis, and semantic analysis. There are two types being Word Tokenisation (splitting text into word) and Sentence Tokenization (splitting text into sentences)
- Stop Words Removal: Involves removing common words that usually do not contribute significant meaning to the text (e.g., "and", "the", "is"). This reduces dimensionality and noise in the data, focusing on more meaningful words.
- Stemming: Reducing words to their root form by removing suffixes (e.g., "running" to "run"). It groups similar words together, simplifying the data and enhancing pattern recognition.
- Lemmatisation: Similar to stemming, but more sophisticated. It reduces words to their base or dictionary form (e.g., "better" to "good"). It provides more accurate linguistic normalization compared to stemming.
- Term Frequency-Inverse Document Frequency (TF-IDF): A statistical measure used to evaluate the importance of a word in a document relative to a collection of documents (corpus). It highlights significant words in documents, improving the accuracy of text analysis. 
- Advanced Tokenization: Tokenization that goes beyond simple splitting, considering context, language nuances, and special cases (e.g., handling contractions, hyphenated words). Provides a more accurate representation of text, especially in complex languages and specialized domains.
- Latent Dirichlet Allocation (LDA): A generative probabilistic model for topic modeling, which discovers the hidden thematic structure in a large collection of documents. It Identifies topics within text data, aiding in content categorization, summarization, and information retrieval.

>- Anon. 2020., states that text processing has the following importance:
- Natural Language Processing (NLP): Enhances machine understanding and generation of human language, enabling applications like chatbots, sentiment analysis, and machine translation.
-	Information Retrieval: Improves search engines and recommendation systems by providing relevant results based on text analysis.
-	Text Mining and Data Mining: Extracts valuable insights and patterns from large textual datasets, aiding decision-making and strategic planning.
-	Sentiment Analysis: Analyses customer reviews, social media posts, and feedback to gauge public opinion and sentiments towards products, services, or events.
-	Document Classification: Automatically categorizes documents into predefined classes, useful in organizing large datasets and automating workflows.

>- According to Durna, M. 2024., The text representation techniques are as follows:
 - Bag-of-Words (BoW): BoW represents text as a collection of words, disregarding grammar and word order but keeping multiplicity. Each document is represented by a vector where each dimension corresponds to a unique word in the corpus, and the value represents the frequency of that word in the document. BoW is simple and effective for tasks like document classification and sentiment analysis. It's computationally efficient but lacks semantic meaning and context.
 - N-grams: N-grams are contiguous sequences of n items (words or characters) in text. They capture local context and allow the model to understand phrases rather than just individual words. Bigrams (n=2) and trigrams (n=3) are commonly used. They are useful in tasks like language modeling, where sequential information is crucial. However, they increase dimensionality and computational complexity.
 - TF-IDF (Term Frequency-Inverse Document Frequency): TF-IDF (Term Frequency-Inverse Document Frequency) is an essential technique in Natural Language Processing (NLP) that addresses the limitations of Bag-of-Words (BoW) models. While BoW models represent words based solely on their frequency within a document, TF-IDF goes further by evaluating the importance of words in a document relative to the entire collection of documents, or corpus. Term Frequency (TF) measures how frequently a term appears within a document relative to the total number of terms in that document. It assigns higher scores to terms that occur frequently within a document. IDF assesses the importance of a term across the entire corpus. It assigns higher scores to terms that are rare across the corpus but occur in specific documents, thus emphasizing their uniqueness and significance. The TF-IDF score for a term is computed by multiplying its TF and IDF scores together. Terms with high TF-IDF scores are deemed important within a specific document context but relatively rare across the entire corpus.
 - Word Embedding: Word2Vec: Word2Vec is a widely used Word Embedding technique that represents words as dense vectors in multi-dimensional space, aiming to capture semantic relationships and contextual nuances. It operates on the premise that words appearing in similar contexts share similar meanings. Word2Vec learns these vector representations by analyzing the words surrounding a target word within a specified context. There are two primary architectures: Continuous Bag of Words (CBOW), which predicts a target word based on its context, and Skip-Gram, which predicts context words given a target word. An illustrative example of Word2Vec's capability is the equation "king - man + woman = queen", demonstrating how vector arithmetic can approximate semantic relationships. This approach leverages the learned word vectors to encode relationships such as gender and royalty. By using cosine similarity, Word2Vec can identify words that are semantically similar to the result of such vector operations, facilitating meaningful analysis of word meanings and associations. Overall, Word2Vec enhances NLP tasks by enabling accurate representation of words and capturing subtle semantic connections within textual data.

>- Text processing faces the following challenges:
-	Ambiguity: Words can have multiple meanings depending on the context (e.g., "bank" can refer to a financial institution or a riverbank).
-	Variability: Different ways of expressing the same concept (e.g., "car", "automobile", "vehicle").
-	Noise: Unnecessary or irrelevant data (e.g., typos, slang, stop words) that can obscure meaningful information.
-	Scalability: Processing large volumes of text data efficiently requires substantial computational resources.

```python
from operator import pos
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator, TransformerMixin
from wordcloud import WordCloud
from collections import Counter
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import networkx as nx


# Step 1: Load the text data
file_path = "C:/Users/lab_services_student/quantum-computing.txt"
with open(file_path, 'r' ) as file:
    text_data = file.read()

# Step 2: Display the first 1000 characters of the text data
print(text_data[:1000])

# Step 3: Download the NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Step 4: 
# This TextPreprocessor class is designed to preprocess text data .
class TextPreprocessor(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.stemmer = PorterStemmer()
        self.lemmatizer = WordNetLemmatizer()
        self.stop_words = set(stopwords.words('english'))

# Fit method is included here to maintain compatibility with scikit-learn pipelines.
    def fit(self, X, y=None):
        return self

# This transform method preprocesses the input text data (X), which is expected to be a list of strings (documents).
    def transform(self, X):
        return [' '.join(self._preprocess(doc)) for doc in X]

# A set of English stopwords from NLTK, which are common words that are often filtered out in text processing tasks because they usually do not carry significant meaning.
    def _preprocess(self, text):
        text = text.lower()
        text = re.sub(r'[^a-z\s]', '', text)
        tokens = word_tokenize(text)
        tokens = [word for word in tokens if word not in self.stop_words]
        stemmed_tokens = [self.stemmer.stem(word) for word in tokens]
        lemmatized_tokens = [self.lemmatizer.lemmatize(word) for word in stemmed_tokens]
        return lemmatized_tokens
    
# __CODE ATTRIBUTION__
# Deepshani. 2024. Text Preprocessing in NLP with Python Codes. Analytics Vidhya. [Online] Available at: https://www.analyticsvidhya.com/blog/2021/06/text-preprocessing-in-nlp-with-python-codes/ [Accessed 26 June 2024]
```

```python
# Step 5: 
# Define the pipeline
pipeline = Pipeline([
    ('preprocessor', TextPreprocessor()),
    ('tfidf', TfidfVectorizer()),
    ('lda', LatentDirichletAllocation(n_components=2, random_state=42))
])

# Fit and transform the data
pipeline.fit([text_data])
lda_matrix_pipeline = pipeline.transform([text_data])

# Function to display the topics
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))

# Display the topics using the pipeline
lda = pipeline.named_steps['lda']
vectorizer = pipeline.named_steps['tfidf']
display_topics(lda, vectorizer.get_feature_names_out(), 10)

# __CODE ATTRIBUTION__
# Whorton, C. 2020. Pipeline For Text Data Pre-processing. Medium. [Online] Available at: https://towardsdatascience.com/pipeline-for-text-data-pre-processing-a9887b4e2db3 [Accessed 26 June 2024]
```

```python
# Step 7: EDA - Data Visualisation 
# A: Word cloud showing the top 10 most significant words

# Function to plot word cloud
def plot_word_cloud(model, feature_names, num_top_words, topic_idx):
    words = [feature_names[i] for i in model.components_[topic_idx].argsort()[:-num_top_words - 1:-1]]
    word_freqs = {word: model.components_[topic_idx][feature_names.tolist().index(word)] for word in words}
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freqs)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"Word cloud showing top 10 most significant words in Topic {topic_idx}")
    plt.show()

# Plot word clouds for each topic
for topic_idx in range(lda.n_components):
    plot_word_cloud(lda, vectorizer.get_feature_names_out(), 10, topic_idx)
    
# __CODE ATTRIBUTION__
# Vu, D. 2023. Generating WordClouds in Python Tutorial. DataCamp. [Online] Available at: https://www.datacamp.com/tutorial/wordcloud-python [Accessed 26 June 2024]
```

# A: Word cloud to visualize the top 10 most significant words in each topic (Exploratory Data Analysis)
   >- Code explanation:
   - The plot_word_cloud function and the subsequent loop that calls it are used to visualize word clouds for each topic extracted by Latent Dirichlet Allocation (LDA).
   - Extracts the top words (num_top_words) for the given topic_idx based on their importance in the LDA model (model.components_).
   - Constructs a dictionary (word_freqs) mapping each word to its weight or importance in the topic.
   - Generates a word cloud (wordcloud) using WordCloud from the wordcloud library, setting parameters for size, background color, and interpolation method.
   - Displays the word cloud using Matplotlib.

```python
# Step 13
# G: Co-occurrence Network graph based on the cosine similarity of terms from the TF-IDF matrix

# Calculate cosine similarity matrix
tfidf_cosine_sim = cosine_similarity(tfidf_matrix.T)

# Convert cosine similarity matrix to DataFrame
cosine_sim_df = pd.DataFrame(tfidf_cosine_sim, index=pipeline.named_steps['tfidf'].get_feature_names_out(), columns=pipeline.named_steps['tfidf'].get_feature_names_out())

# Network graph
G = nx.Graph()

# Add edges based on cosine similarity threshold 
threshold = 0.2
edges = []
for i, row in cosine_sim_df.iterrows():
    for j, sim in row.items():
        if i != j and sim > threshold:
            edges.append((i, j, sim))

G.add_weighted_edges_from(edges)

# Calculate the layout positions for nodes using spring layout
pos = nx.spring_layout(G, k=0.15, iterations=20)

# Plot the network graph
plt.figure(figsize=(15, 12))
nodes = nx.draw_networkx_nodes(G, pos, node_size=300, node_color='orange', alpha=0.6)
edges = nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.7, edge_color='gray')
labels = nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif')

# Add title and color bar
plt.title('Co-occurrence Network Graph of Terms', fontsize=15)
plt.colorbar(nodes)
plt.axis('off')
plt.show()

# __CODE ATTRIBUTION__
# Anon. n.d. Network Graphs in Python. Plotly. [Online] Available at: https://plotly.com/python/network-graphs/ [Accessed 27 June 2024]
# Ito, E. 2018. Simple Co-occurrence Network. Kaggle. [Online] Available at: https://www.kaggle.com/code/itoeiji/simple-co-occurrence-network [Accessed 27 June 2024]
```
